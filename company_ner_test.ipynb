```python
import spacy
```

```python
nlp = spacy.load('en')
f   = open('./news.data','r')

NASDAQ_PATH = "./NASDAQ.txt"
NYSE_PATH   = "./NYSE.txt"
def get_company_dictionary(path, name2symbol, symbol2name):
    
    
    with open(path, 'r') as f:
        
        for d in f:
            d = d.strip()
            data = d.split("\t")
            
            #print(data)
            #print(data[0], data[1])
            # ticker
            symbol2name[data[0]] = data[1]
            name2symbol[data[1]] = data[0]
            
            
name2symbol = {}
symbol2name = {}
get_company_dictionary(NASDAQ_PATH, name2symbol, symbol2name)
get_company_dictionary(NYSE_PATH,   name2symbol, symbol2name)
```
